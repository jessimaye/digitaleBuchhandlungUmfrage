```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv("umfrage_ergebnisse.csv", encoding="utf-16")
```

```python
#Fälle löschen, bei denen kein Alter angegeben wurde, da diese nicht vollständig ausgefüllt wurden 
results.dropna(subset=["SD02_01"], inplace=True)
```

```python
#Nur Fälle behalten, die zur Zielgruppe gehören (Alter: 18-39)
results_targetgroup = results[results["SD02_01"]<40]
```

```python
#Alle Teilnehmende, die die Umfrage vollständig ausgefüllt haben
len(results)
```

```python
#Alle Teilnehmende, die die Umfrage vollständig ausgefüllt haben und zur Zielgruppe gehören (Alter:18-39)
len(results_targetgroup)
```

```python
#Verteilung Geschlecht innerhalb der Zielgruppe, 1=weiblich, 2=männlich, 3=divers, 4=keine Angabe
results_targetgroup["SD05"].value_counts()
```

```python
#Verteilung Alter innerhalb der Zielgruppe
results_targetgroup["SD02_01"].describe()
```

```python
#Verteilung technisches Hintergrundwissen innerhalb der Zielgruppe
#1=sehr schlecht, 2=schlecht, 3=mittel, 4=gut, 5=sehr gut
results_targetgroup["SD06"].value_counts()
```

```python
#Wie sind die Teilnehmenden auf das Buch gekommen, das sie zum Zeitpunkt der Umfrage gelesen haben
#1=Empfehlung durch Freunde, Familie oder Bekannte, 2=Soziale Netzwerke, 3=Medien, 4=Buchhandlung, 5=Sonstiges, 6=Bestsellerliste
results_targetgroup["L102"].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(14,6))
ax.bar(["Freunde/Bekannte/Familie", "Soziale Netzwerke", "Medien", "Buchhandlung", "Bestsellerliste", "Sonstiges"], [14,7,5,6,2,11], width=0.5)

ax.set_title("Wie sind Sie auf Ihr aktuelles Buch gekommen?")
ax.set_xlabel("Antwortmöglichkeiten")
ax.set_ylabel("Häufigkeit der Nennungen")

plt.show()
```

```python
#Nutzen Sie digitale Buchempfehlungen?
#1=Ja, 2=Nein
results_targetgroup["E101"].value_counts()
```

```python
#Falls ja: Wie zufrieden sind sie mit diesen Empfehlungen?
#1=sehr unzufrieden, 2=unzufrieden, 3=neutral, 4=zufrieden, 5=sehr zufrieden, -9=keine Angabe
results_targetgroup["E102"].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.bar(["sehr unzufrieden", "unzufrieden", "neutral", "zufrieden", "sehr zufrieden"], [2,2,11,10,3], width=0.3)

ax.set_title("Wie zufrieden sind Sie mit digitalen Buchempfehlungen?")
ax.set_xlabel("Antwortmöglichkeiten")
ax.set_ylabel("Häufigkeit der Nennungen")

plt.show()
```
